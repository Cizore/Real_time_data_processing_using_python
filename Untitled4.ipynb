server code to generate random ethernet data and send it to clients


```python
import socket
import random
import time
import threading

# Function to generate random Ethernet data
def generate_random_ethernet_data():
    data_length = random.randint(20, 1500)  # Ethernet frame size can vary from 20 to 1500 bytes
    return bytes(random.getrandbits(8) for _ in range(data_length))

# Server code to send random Ethernet data to client continuously
def start_server(host='127.0.0.1', port=65417):
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((host, port))
    server_socket.listen()
    print(f"Server listening on {host}:{port}")
    
    def handle_client_connection(client_socket):
        try:
            while True:
                data = generate_random_ethernet_data()
                client_socket.sendall(data)
                time.sleep(0.1)  # Adjust the sleep time as needed to control the data sending rate
        except Exception as e:
            print(f"Connection error: {e}")
        finally:
            client_socket.close()
    
    while True:
        client_socket, addr = server_socket.accept()
        print(f"Accepted connection from {addr}")
        client_handler = threading.Thread(target=handle_client_connection, args=(client_socket,))
        client_handler.start()

# Start the server in a separate thread to keep the notebook interactive
server_thread = threading.Thread(target=start_server)
server_thread.daemon = True
server_thread.start()
```

```python
import socket
import struct
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier  # Import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import time
import IPython.display as display

# Function to extract Ethernet features from raw data
def extract_ethernet_features(data):
    if len(data) < 14:
        return None
    dest_mac = data[0:6].hex()
    src_mac = data[6:12].hex()
    ethertype = struct.unpack('!H', data[12:14])[0]
    payload = data[14:]
    return {
        'dest_mac': dest_mac,
        'src_mac': src_mac,
        'ethertype': ethertype,
        'payload_length': len(payload),
        'payload': payload
    }

# Function to preprocess features
def preprocess_features(features, le_dest=None, le_src=None, scaler=None, seen_dest_macs=None, seen_src_macs=None):
    dest_macs = [f['dest_mac'] for f in features]
    src_macs = [f['src_mac'] for f in features]
    ethertypes = [f['ethertype'] for f in features]
    payload_lengths = [f['payload_length'] for f in features]
    
    if seen_dest_macs is None:
        seen_dest_macs = set()
    if seen_src_macs is None:
        seen_src_macs = set()
    
    new_dest_macs = set(dest_macs) - seen_dest_macs
    new_src_macs = set(src_macs) - seen_src_macs
    
    if le_dest is None:
        le_dest = LabelEncoder()
        le_dest.fit(list(seen_dest_macs) + list(new_dest_macs))
    else:
        if new_dest_macs:
            le_dest.fit(list(le_dest.classes_) + list(new_dest_macs))
    
    if le_src is None:
        le_src = LabelEncoder()
        le_src.fit(list(seen_src_macs) + list(new_src_macs))
    else:
        if new_src_macs:
            le_src.fit(list(le_src.classes_) + list(new_src_macs))
    
    dest_macs_encoded = le_dest.transform(dest_macs)
    src_macs_encoded = le_src.transform(src_macs)
    
    feature_array = np.array(list(zip(dest_macs_encoded, src_macs_encoded, ethertypes, payload_lengths)))
    
    if scaler is None:
        scaler = StandardScaler()
        feature_array_scaled = scaler.fit_transform(feature_array)
    else:
        feature_array_scaled = scaler.transform(feature_array)
    
    seen_dest_macs.update(new_dest_macs)
    seen_src_macs.update(new_src_macs)
    
    return feature_array_scaled, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs

# Function to display processed Ethernet features as a DataFrame
def display_processed_features_as_df(features):
    df = pd.DataFrame(features, columns=['dest_mac', 'src_mac', 'ethertype', 'payload_length'])
    display.clear_output(wait=True)
    display.display(df)

# Function to plot accuracy vs time
def plot_accuracy_vs_time(timestamps, accuracies, max_display_seconds=60):
    plt.figure(figsize=(10, 5))
    
    # Limit display to the last `max_display_seconds` seconds
    display_cutoff = datetime.now() - timedelta(seconds=max_display_seconds)
    recent_timestamps = [ts for ts in timestamps if ts > display_cutoff]
    recent_accuracies = accuracies[-len(recent_timestamps):]
    
    plt.plot(recent_timestamps, recent_accuracies, marker='o')
    plt.xlabel('Time')
    plt.ylabel('Accuracy')
    plt.title('Accuracy vs Time')
    plt.grid(True)
    plt.xticks(rotation=45)
    
    plt.show()
    plt.pause(0.1)  # Pause to allow the plot to be displayed
    
    # Clear the plot after a delay (e.g., 3 seconds)
    time.sleep(3)
    plt.clf()

# Function to receive Ethernet data from server and apply various classifiers
def receive_and_process_ethernet_data(host='127.0.0.1', port=65417, classifier='SGD'):
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect((host, port))
    print(f"Connected to server at {host}:{port}")

    collected_features = []
    if classifier == 'SGD':
        model = SGDClassifier(loss='log')
    elif classifier == 'RandomForest':
        model = RandomForestClassifier()  # Initialize RandomForestClassifier
    else:
        raise ValueError(f"Unknown classifier: {classifier}")
    
    le_dest, le_src, scaler = None, None, None
    seen_dest_macs, seen_src_macs = set(), set()
    test_size = 100  # Number of samples to use for the test set
    timestamps, accuracies = [], []

    try:
        while True:
            data = client_socket.recv(1500)  # Adjust buffer size as needed
            if not data:
                break
            
            # Extract and preprocess features
            features = extract_ethernet_features(data)
            if features:
                collected_features.append(features)
                if len(collected_features) > 1:  # Ensure there's enough data to display
                    processed_features, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs = preprocess_features(
                        collected_features, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs)
                    display_processed_features_as_df(processed_features)

                    # Split data into training and test sets
                    if len(collected_features) > test_size:
                        X_train = processed_features[:-test_size]
                        X_test = processed_features[-test_size:]
                        y_train = np.random.randint(0, 2, size=X_train.shape[0])  # Simulated binary target variable
                        y_test = np.random.randint(0, 2, size=X_test.shape[0])  # Simulated binary target variable

                        if classifier == 'SGD':
                            # Train SGDClassifier incrementally
                            model.partial_fit(X_train, y_train, classes=[0, 1])
                        else:
                            # Train RandomForestClassifier on the entire dataset
                            model.fit(X_train, y_train)

                        # Predict and evaluate model
                        y_pred = model.predict(X_test)
                        accuracy = accuracy_score(y_test, y_pred)
                        print(f"Accuracy: {accuracy}")

                        # Record timestamp and accuracy
                        timestamps.append(datetime.now())
                        accuracies.append(accuracy)

                        # Plot accuracy vs time
                        plot_accuracy_vs_time(timestamps, accuracies)

    except Exception as e:
        print(f"Connection error: {e}")
    finally:
        client_socket.close()

# Start receiving and processing data from the server with SGDClassifier by default
receive_and_process_ethernet_data(classifier='RandomForest')  # Change classifier to 'RandomForest' to use RandomForestClassifier
```

```python
import socket
import struct
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from datetime import datetime, timedelta
import time
import IPython.display as display

# Function to extract Ethernet features from raw data
def extract_ethernet_features(data):
    if len(data) < 14:
        return None
    dest_mac = data[0:6].hex()
    src_mac = data[6:12].hex()
    ethertype = struct.unpack('!H', data[12:14])[0]
    payload = data[14:]
    return {
        'dest_mac': dest_mac,
        'src_mac': src_mac,
        'ethertype': ethertype,
        'payload_length': len(payload),
        'payload': payload
    }

# Function to preprocess features
def preprocess_features(features, le_dest=None, le_src=None, scaler=None, seen_dest_macs=None, seen_src_macs=None):
    dest_macs = [f['dest_mac'] for f in features]
    src_macs = [f['src_mac'] for f in features]
    ethertypes = [f['ethertype'] for f in features]
    payload_lengths = [f['payload_length'] for f in features]
    
    if seen_dest_macs is None:
        seen_dest_macs = set()
    if seen_src_macs is None:
        seen_src_macs = set()
    
    new_dest_macs = set(dest_macs) - seen_dest_macs
    new_src_macs = set(src_macs) - seen_src_macs
    
    if le_dest is None:
        le_dest = LabelEncoder()
        le_dest.fit(list(seen_dest_macs) + list(new_dest_macs))
    else:
        if new_dest_macs:
            le_dest.fit(list(le_dest.classes_) + list(new_dest_macs))
    
    if le_src is None:
        le_src = LabelEncoder()
        le_src.fit(list(seen_src_macs) + list(new_src_macs))
    else:
        if new_src_macs:
            le_src.fit(list(le_src.classes_) + list(new_src_macs))
    
    dest_macs_encoded = le_dest.transform(dest_macs)
    src_macs_encoded = le_src.transform(src_macs)
    
    feature_array = np.array(list(zip(dest_macs_encoded, src_macs_encoded, ethertypes, payload_lengths)))
    
    if scaler is None:
        scaler = StandardScaler()
        feature_array_scaled = scaler.fit_transform(feature_array)
    else:
        feature_array_scaled = scaler.transform(feature_array)
    
    seen_dest_macs.update(new_dest_macs)
    seen_src_macs.update(new_src_macs)
    
    return feature_array_scaled, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs

# Function to display processed Ethernet features as a DataFrame
def display_processed_features_as_df(features):
    df = pd.DataFrame(features, columns=['dest_mac', 'src_mac', 'ethertype', 'payload_length'])
    display.clear_output(wait=True)
    display.display(df)

# Function to plot accuracy vs time with volume bars
def plot_accuracy_vs_time(timestamps, accuracies, volumes, max_display_seconds=60):
    plt.figure(figsize=(14, 10))

    display_cutoff = datetime.now() - timedelta(seconds=max_display_seconds)
    recent_timestamps = [ts for ts in timestamps if ts > display_cutoff]
    recent_accuracies = accuracies[-len(recent_timestamps):]
    recent_volumes = volumes[-len(recent_timestamps):]

    sns.set(style="whitegrid")

    ax1 = plt.subplot2grid((4, 4), (0, 0), rowspan=3, colspan=4)
    ax2 = plt.subplot2grid((4, 4), (3, 0), rowspan=1, colspan=4, sharex=ax1)

    ax1.plot(recent_timestamps, recent_accuracies, marker='o', color='mediumvioletred', linewidth=3.5, markersize=12)
    ax1.fill_between(recent_timestamps, recent_accuracies, color='palegreen', alpha=0.5)

    ax1.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.8)
    ax1.set_ylabel('Accuracy', fontsize=16, fontweight='bold', color='darkblue')
    ax1.set_title('Real-Time Accuracy vs Time', fontsize=20, fontweight='bold', color='purple')

    ax1.xaxis.set_major_locator(mdates.SecondLocator(interval=5))
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right', fontsize=14, color='darkgreen')
    ax1.tick_params(axis='y', labelsize=14, colors='darkgreen')

    if recent_timestamps and recent_accuracies:
        ax1.annotate(f'{recent_accuracies[-1]:.2f}', 
                     xy=(recent_timestamps[-1], recent_accuracies[-1]), 
                     xytext=(recent_timestamps[-1] + timedelta(seconds=5), recent_accuracies[-1] + 0.02),
                     arrowprops=dict(facecolor='purple', shrink=0.05, width=2, headwidth=8),
                     fontsize=14, fontweight='bold', color='red')

    ax1.legend(['Accuracy over Time'], loc='upper left', fontsize=14, frameon=True, shadow=True, facecolor='lightblue')

    ax2.bar(recent_timestamps, recent_volumes, width=0.01, color='steelblue')
    ax2.set_ylabel('Volume', fontsize=16, fontweight='bold', color='darkblue')
    ax2.tick_params(axis='y', labelsize=14, colors='darkgreen')
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right', fontsize=14, color='darkgreen')

    background_gradient = np.linspace(0, 1, 100).reshape(1, -1)
    ax1.imshow(background_gradient, aspect='auto', cmap='viridis', alpha=0.2, extent=[mdates.date2num(recent_timestamps[0]), mdates.date2num(recent_timestamps[-1]), 0, 1])

    plt.show()
    plt.pause(0.001)
    time.sleep(0.1)
    plt.clf()

# Function to receive Ethernet data from server and apply Logistic Regression
def receive_and_process_ethernet_data(host='127.0.0.1', port=65417):
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect((host, port))
    print(f"Connected to server at {host}:{port}")

    collected_features = []
    model = SGDClassifier(loss='log')
    le_dest, le_src, scaler = None, None, None
    seen_dest_macs, seen_src_macs = set(), set()
    test_size = 100
    timestamps, accuracies, volumes = [], [], []
    volume_counter = 0

    try:
        while True:
            data = client_socket.recv(1500)
            if not data:
                break
            
            features = extract_ethernet_features(data)
            if features:
                collected_features.append(features)
                volume_counter += 1
                if len(collected_features) > 1:
                    processed_features, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs = preprocess_features(
                        collected_features, le_dest, le_src, scaler, seen_dest_macs, seen_src_macs)
                    display_processed_features_as_df(processed_features)

                    if len(collected_features) > test_size:
                        X_train = processed_features[:-test_size]
                        X_test = processed_features[-test_size:]
                        y_train = np.random.randint(0, 2, size=X_train.shape[0])
                        y_test = np.random.randint(0, 2, size=X_test.shape[0])

                        model.partial_fit(X_train, y_train, classes=[0, 1])
                        y_pred = model.predict(X_test)
                        accuracy = accuracy_score(y_test, y_pred)
                        print(f"Accuracy: {accuracy}")

                        timestamps.append(datetime.now())
                        accuracies.append(accuracy)
                        volumes.append(volume_counter)
                        volume_counter = 0

                        plot_accuracy_vs_time(timestamps, accuracies, volumes)

    except Exception as e:
        print(f"Connection error: {e}")
    finally:
        client_socket.close()

receive_and_process_ethernet_data()
```
